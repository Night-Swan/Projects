```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example 1: Count of passengers by survival status
sns.countplot(x='Survived', data=df)
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.title('Count of Survivors vs Non-Survivors')
plt.show()

# Example 2: Count of passengers by passenger class (Pclass)
sns.countplot(x='Pclass', data=df)
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Count')
plt.title('Count of Passengers by Class')
plt.show()

# Example 3: Count of passengers by gender (Sex)
sns.countplot(x='Sex', data=df)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count of Passengers by Gender')
plt.show()

# Example 4: Count of passengers by embarkation port (Embarked)
sns.countplot(x='Embarked', data=df)
plt.xlabel('Embarkation Port')
plt.ylabel('Count')
plt.title('Count of Passengers by Embarkation Port')
plt.show()
```

```python
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
df = df.drop('Name', axis=1)
df.head()
```

```python
df['Cabin_Letter'] = df['Cabin'].str[0].where(df['Cabin'].notna())
df.head()
```

```python
sns.countplot(data=df, x='Cabin_Letter', hue='Survived')
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] 
df.head()
```

```python
sns.countplot(data=df, x='FamilySize', hue='Survived')
```
